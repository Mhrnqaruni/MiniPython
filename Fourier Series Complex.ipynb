```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Define the variable and the function
x = sp.symbols('x')
f = x**3

# Define the limits
L = sp.pi  # You can adjust this value according to your function's period
k = sp.symbols('k', integer=True)

# Calculate the Fourier coefficients
Ck = (1/(2*L))*sp.integrate(f*sp.exp(-sp.I*k*sp.pi*x/L), (x, -L, L))

# Define the number of terms in the Fourier series
N = 10  # You can adjust this value

# Calculate the Fourier series
fourier_series = sum(Ck.subs(k, n)*sp.exp(sp.I*n*sp.pi*x/L) for n in range(-N, N+1))

# Lambdify the Fourier series for plotting
fourier_func = sp.lambdify(x, fourier_series, 'numpy')

# Generate x values
x_vals = np.linspace(float(-L), float(L), 400)

# Generate y values (Fourier series)
y_vals = fourier_func(x_vals)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='Fourier series of $x^2$')
plt.legend()
plt.grid(True)
plt.show()
```
